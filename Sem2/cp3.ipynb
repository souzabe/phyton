```python
#Guardando os preços
precos_quartos = {
    'simples': 100,
    'duplo': 150,
    'suite': 200
}

reservas = []

#Função para adicionar reserva
def adicionar_reserva(nome, tipo_quarto, duracao):
    if tipo_quarto not in precos_quartos:
        print("Tipo de quarto inválido.")
        return
    reserva = {
        'nome': nome,
        'tipo_quarto': tipo_quarto,
        'duracao': duracao,
        'custo_total': precos_quartos[tipo_quarto] * duracao
    }
    reservas.append(reserva)
    print(f"Reserva adicionada para {nome}, tipo: {tipo_quarto}, duração: {duracao} noites.")

#Função para remover reserva
def remover_reserva(nome):
    global reservas
    reservas = [reserva for reserva in reservas if reserva['nome'] != nome]
    print(f"Reserva de {nome} removida.")

#Função para listar as reservas
def listar_reservas():
    if not reservas:
        print("Não há reservas.")
    else:
        for reserva in reservas:
            print(f"Hóspede: {reserva['nome']}, Tipo: {reserva['tipo_quarto']}, Duração: {reserva['duracao']} noites, Custo total: R${reserva['custo_total']}")


adicionar_reserva("Maria", "suite", 3)
adicionar_reserva("João", "duplo", 2)
listar_reservas()
remover_reserva("Maria")
listar_reservas()
```

```python
# Lista para guardar os carros estacionados
estacionamento = []

# Função para adicionar um carro
def adicionar_carro(marca, cor, placa):
    carro = {
        'marca': marca,
        'cor': cor,
        'placa': placa
    }
    estacionamento.append(carro)
    print(f"Carro {marca} ({cor}) com placa {placa} foi adicionado.")

# Função para remover um carro
def remover_carro(placa):
    global estacionamento
    estacionamento = [carro for carro in estacionamento if carro['placa'] != placa]
    print(f"Carro com placa {placa} foi removido..")

# Função para listar os carros
def listar_carros():
    if not estacionamento:
        print("Estacionamento está vazio, tapado.")
    else:
        for carro in estacionamento:
            print(f"Marca: {carro['marca']}, Cor: {carro['cor']}, Placa: {carro['placa']}")

# Simulação do sistema com loop
while True:
    print("\nEscolha uma opção:")
    print("1 - Adicionar carro")
    print("2 - Remover carro")
    print("3 - Listar carros")
    print("4 - Sair")
    
    escolha = input("Opção: ")
    
    if escolha == '1':
        marca = input("Marca do carro: ")
        cor = input("Cor do carro: ")
        placa = input("Placa do carro: ")
        adicionar_carro(marca, cor, placa)
    
    elif escolha == '2':
        placa = input("Placa do carro a remover: ")
        remover_carro(placa)
    
    elif escolha == '3':
        listar_carros()
    
    elif escolha == '4':
        print("Sistema encerrado.")
        break
    
    else:
        print("Opção inválida! Tenta de novo.")
```

```python
# Dicionário para guardar os projetos e suas tarefas
projetos = {}

# Função para adicionar tarefa em um projeto
def adicionar_tarefa(projeto, titulo, descricao, responsavel, status='não iniciado'):
    if projeto not in projetos:
        projetos[projeto] = []
    tarefa = {
        'titulo': titulo,
        'descricao': descricao,
        'responsavel': responsavel,
        'status': status
    }
    projetos[projeto].append(tarefa)
    print(f"Tarefa '{titulo}' adicionada ao projeto '{projeto}'.")

# Função para atualizar uma tarefa
def atualizar_tarefa(projeto, titulo, status=None, descricao=None, responsavel=None):
    if projeto in projetos:
        for tarefa in projetos[projeto]:
            if tarefa['titulo'] == titulo:
                if status:
                    tarefa['status'] = status
                if descricao:
                    tarefa['descricao'] = descricao
                if responsavel:
                    tarefa['responsavel'] = responsavel
                print(f"Tarefa '{titulo}' atualizada no projeto '{projeto}'.")
                return
        print(f"Tarefa '{titulo}' não encontrada no projeto '{projeto}', seu pastel.")
    else:
        print(f"Projeto '{projeto}' não existe, acéfalo.")

# Função para listar tarefas de um projeto
def listar_tarefas(projeto):
    if projeto in projetos and projetos[projeto]:
        print(f"\nTarefas do projeto '{projeto}':")
        for tarefa in projetos[projeto]:
            print(f" - Título: {tarefa['titulo']}")
            print(f"   Descrição: {tarefa['descricao']}")
            print(f"   Responsável: {tarefa['responsavel']}")
            print(f"   Status: {tarefa['status']}\n")
    else:
        print(f"Não há tarefas no projeto '{projeto}' ou o projeto não existe.")

# Loop para simular o sistema
while True:
    print("\nEscolha uma opção:")
    print("1 - Adicionar tarefa")
    print("2 - Atualizar tarefa")
    print("3 - Listar tarefas de um projeto")
    print("4 - Sair")
    
    opcao = input("Opção: ")
    
    if opcao == '1':
        projeto = input("Nome do projeto: ")
        titulo = input("Título da tarefa: ")
        descricao = input("Descrição da tarefa: ")
        responsavel = input("Responsável pela tarefa: ")
        adicionar_tarefa(projeto, titulo, descricao, responsavel)
    
    elif opcao == '2':
        projeto = input("Nome do projeto: ")
        titulo = input("Título da tarefa a atualizar: ")
        status = input("Novo status (ou deixe vazio): ")
        descricao = input("Nova descrição (ou deixe vazio): ")
        responsavel = input("Novo responsável (ou deixe vazio): ")
        atualizar_tarefa(projeto, titulo, status if status else None, descricao if descricao else None, responsavel if responsavel else None)
    
    elif opcao == '3':
        projeto = input("Nome do projeto: ")
        listar_tarefas(projeto)
    
    elif opcao == '4':
        print("Saindo do sistema!")
        break
    
    else:
        print("Opção inválida! Tenta de novo.")
```

```python
# Função para solicitar dados e calcular o bônus
def solicitar_dados():
    nome = input("Nome do funcionário: ")
    cpf = input("CPF do funcionário: ")
    salario_mes1 = float(input("Salário do primeiro mês: "))
    salario_mes2 = float(input("Salário do segundo mês: "))
    salario_mes3 = float(input("Salário do terceiro mês: "))
    
    # Calcula a média salarial
    salario_medio = (salario_mes1 + salario_mes2 + salario_mes3) / 3
    
    # Calcula o bônus de acordo com a média
    if salario_medio > 10000:
        bonus = salario_medio * 0.15
    elif salario_medio >= 5000:
        bonus = salario_medio * 0.10
    else:
        bonus = salario_medio * 0.07

    # Exibe o resultado
    print(f"\nFuncionário: {nome}")
    print(f"CPF: {cpf}")
    print(f"Salário médio: R${salario_medio:.2f}")
    print(f"Bônus: R${bonus:.2f}")

# Chama a função para executar o programa
solicitar_dados()
```
